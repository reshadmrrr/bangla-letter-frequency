```python
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt 
import numpy as np 

freq = {}

def readFile(input = "input.txt"):
    f = open(input, "r", encoding="utf-8")
    for char in f.read():
        if(char in freq):
            freq[char] += 1
        else:
            freq[char] = 1


def main():
    readFile()
    alphabets = 'অআ াই িঈ ীউ ুঊ ূঋ ৃএ েঐ ৈও োঔ ৌকখগঘঙচছজঝঞটঠডঢণতথদধনপফবভমযরলশষসহড়ঢ়য়ৎংঃঁ'
    alphabets ="".join(alphabets.split()) 
    counts = []
    for char in alphabets:
        if char in freq:
            counts.append(freq[char])
        else:
            counts.append(0)
    count = 0
    for val in counts:
        count += val
    counts = [100 * val / count for val in counts]

    fig, ax = plt.subplots(figsize=(40, 20))
    prop = fm.FontProperties(fname="Shorif Jonota.ttf")
    ax.legend(prop=prop)

    font_dirs = ["./",]
    font_files = fm.findSystemFonts(fontpaths=font_dirs)
    font_list = fm.createFontList(font_files)
    fm.fontManager.ttflist.extend(font_list)
    plt.tick_params(labelsize=20)

    plt.bar(range(len(alphabets)), counts, align='center')
    plt.xticks(np.arange(len(alphabets)),
               list(alphabets), fontfamily='Shorif Jonota', fontsize=26)

    plt.xlabel('বর্ণ-সমূহ', fontsize=32, fontfamily='Shorif Jonota')
    plt.ylabel('শতকরা-হার (%)', fontsize=32, fontfamily='Shorif Jonota')

    fig.suptitle(
        'Relative Frequencies of letters in Bengali text\nby Monayam Reshad, ID#16701080\nref: https://bn.wikipedia.org/wiki/বাংলাদেশের_ইতিহাস', fontsize=26)

    plt.show()
if __name__ == "__main__":
    main()
```
